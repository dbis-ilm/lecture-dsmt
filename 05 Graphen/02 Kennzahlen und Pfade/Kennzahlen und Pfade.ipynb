## Pfade und Zusammenhang
Ein **Pfad**, manchmal auch als Weg bezeichnet, ist eine Abfolge von Knoten, wobei zwei aufeinanderfolgende Knoten immer durch eine Kante verbunden sind. Im nachfolgenden Beispiel existiert also der Pfad $A, B, D$, nicht aber der Pfad $A, C, D$, da zwischen $A$ und $C$ keine Kante existiert.



In ungerichteten Graphen sind Pfade immer symmetrisch. Die Existenz des Pfades $A, B, D$ impliziert also auch die Existenz des Pfades $D, B, A$.

Die **Länge** eines Pfads bestimmt sich dabei aus der Anzahl an Kanten, die dem Pfad angehören. Ein nichttrivialer Pfad hat also eine minimale Länge von $1$.

In komplexen Graphen gibt es dabei häufig mehrere Wege zu Ziel. Von $A$ lässt sich $C$ beispielsweise mit Hilfe der Pfade $A, B, D, C$ und $A, B, C$ erreichen. Ein **kürzester Pfad** zwischen zwei Knoten ist genau die Teilmenge aller Pfade zwischen ebendiesen Knoten, welche die minimale Länge besitzen. (Es kann auch mehrere kürzeste Pfade geben!)

Ein Graph heißt **zusammenhängend**, wenn von jedem Knoten ein Pfad zu jedem anderen Knoten im Graphen existiert. Im nachfolgenden Beispiel existiert beispielsweise kein Pfad von $A$ zu $E$.



Graphen, die nicht zusammenhängend sind, zerfallen in sogenannte **Komponenten**, die jeweils für sich zusammenhängend sind.

## Einfache Kennzahlen

### Dichte
Die Dichte eines Graphen gibt relativ an, wie viele der maximal möglichen Kanten tatsächlich vorhanden sind. Ein Wert von $1$ entspricht einem vollständig verbundenen Graphen, während $0$ einen Graphen ohne Kanten widerspiegelt. Die Dichte wird wie folgt berechnet:

$$
d(G) = \dfrac{\left|E\right|}{\binom{\left|V\right|}{2}}
$$

### Durchmesser
Der Durchmesser eines Graphen entspricht der Länge des längsten kürzesten Pfades.



Im vorangegangenen Beispiel befindet sich der längste kürzeste Pfad zwischen `A` und `D` und besitzt die Länge $3$. Es existiert außerdem ein Pfad zwischen diesen beiden Knoten, der durch einen Umweg über `E` länger, aber kein kürzester Pfad ist. Alle anderen Pfade sind kürzer.

### Knotengrad
Der Knotengrad ist eine der einfachsten Metriken zur Einschätzung der Relevanz und Eingebundenheit einzelner Knoten. Er gibt für jeden Knoten einzeln an, wie viele Kanten von ihm ausgehen.



Knoten mit einem hohen Grad besitzen also viele Verbindungen und sind somit besser innerhalb des Graphen vernetzt.

Innerhalb eines gerichteten Graphen kann außerdem zwischen dem eingehenden (*in-degree*) und ausgehenden (*out-degree*) Knotengrad unterschieden werden.

### Bäume
Ein Baum ist ein zusammenhängender, kreisfreier, ungerichteter Graph. Als Wurzel kann dabei ein Knoten ausgewiesenen werden, um dem Einstiegspunkt der zugehörigen Datenstruktur gerecht zu werden. Knoten, die lediglich den Knotengrad $1$ besitzen, werden Blatt genannt. Alle verbleibenden Knoten werden als innere Knoten bezeichnet.

### Bipartite Graphen
Die Knotenmenge bipartiter Graphen lassen sich in zwei Teilmengen zerlegen, sodass Kanten ausschließlich zwischen diesen Teilmengen verlaufen. Innerhalb beider Teilmengen existieren dagegen keine Kanten mehr. Mit einer geeigneten Darstellung lässt sich das Prinzip bipartiter Graphen einfach veranschaulichen.



Bipartite Graphen werden beispielsweise verwendet, um Zuordnungsprobleme zu modellieren.
