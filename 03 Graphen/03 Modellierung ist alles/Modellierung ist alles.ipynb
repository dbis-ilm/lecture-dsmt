## Das Königsberger Brückenproblem
Königsberg (heute Kaliningrad) ist nicht nur für die nach dieser Stadt benannten Klopse und den einzigen ganzjährig eisfreien Hafen Russlands an der Ostsee bekannt, sondern ging im frühen 18. Jahrhundert auch in die Geschichte der Mathematik ein. Die damals noch in Ostpreußen gelegene Stadt wird geteilt vom Fluss Pregel, der in zwei Armen zwei Inseln innerhalb der Stadt umschließt. Zum Überqueren des Flusses standen sieben Brücken zur Verfügung.

 <br>
[Das Königsberger Brückenproblem illustriert von Bogdan Giuşcă (CC BY-SA 3.0)](https://commons.wikimedia.org/wiki/File:Konigsberg_bridges.png)

Für eine Parade sollte nun ein Weg gefunden werden, der jede dieser Brücken exakt einmal überschreitete und am Ausgangspunkt endete. Leonhard Euler erkannte, dass dieses Problem nicht davon abhing, wie die Insel- und Festlandstücken angeordnet sind. Stattdessen beschränkte er seinen Blick auf die Brücken, also die Kanten untereinander, und "löste" damit das Problem.



Der nach ihm benannte Eulerkreis ist deshalb ein Kantenzug, dessen Start- und Zielknoten identisch sind und der alle Kanten eines Graphen genau einmal durchläuft. Ein solcher existiert in einem ungerichteten, zusammenhängenden Graphen genau dann, wenn keiner seiner Knoten einen ungeraden Knotengrad besitzen. Die Parade konnte demnach nicht wie geplant durchgeführt werden, während der Satz aber als erster der Graphentheorie in die Geschichte der Mathematik einging.

Ein verwandtes Problem ist der Eulerweg, der als Einschränkung gegenüber dem Eulerkreis besitzt, dass Start- und Endpunkt nicht identisch sein müssen. Das Haus vom Nikolaus ist daher nicht nur ein Rätsel für Kinder, sondern ebenfalls eines für Mathematiker.

## Bauer, Kohl, Schaf, Wolf
Ein Bauer befindet sich mit einem Kohlkopf, einem Schaf und einem Wolf am Ufer eines Flusses. Er möchte mit allen übersetzen und hat dazu leider nur sein viel zu kleines Ruderboot zur Verfügung, das neben ihm nur ein weiteres Objekt befördert werden kann. Er muss also mehrfach in beide Richtungen den Fluss überqueren, um alles zu befördern.

Der Bauer muss aber natürlich achtsam sein, wen er unbeaufsichtigt zurücklässt. Lässt er Schaf und Kohlkopf allein, frisst sich das Schaf satt und der Bauer hat kein Abendbrot mehr. Lässt er Schaf und Wolf allein, lässt er ebenfalls der Natur seinen Lauf und wird im Winter frieren. Nur solange er selbst anwesend ist, kann er solche Umtriebe verhindern.

Welche Fahrten muss der Bauer also unternehmen, um **sicher** alle Tiere, Pflanzen und sich selbst auf die andere Seite zu befördern?

### Knoten
Zuerst sollen die Knoten modelliert werden. Als Benennung verwenden wir das folgende Schema: Der Fluss wird durch einen vertikalen Strich gekennzeichnet. Die jeweiligen Akteure werden durch ihre Anfangsbuchstaben identifiziert und befinden sich entweder links oder rechts vom Fluss. `BKSW|` ist also die Ausgangssituation (alle Akteure befinden sich links des Flusses), während `|BKSW` (alle Akteure befinden sich auf der rechten Seite des Flusses) die Zielsituation beschreibt.

Abgebildet werden ausschließlich die gültigen Zustände. So ist beispielsweise `BS|KW` legitim, da der Wolf keinen Kohl frisst. Einen ebenfalls gültigen Zustand stellt `BSW|K` dar, da der Bauer auf der linken Seite auf das Schaf aufpasst und auf der rechten Seite niemand am Kohl interessiert ist. `SW|BK` ist dagegen kein gültiger Zustand und wird demnach nicht als Knoten in den Graphen aufgenommen.

### Kanten
Eine Kante wird nun zwischen zwei Knoten eingefügt, wenn zwei Zustände durch eine Überfahrt ineinander überführt werden können. Also entweder wechselt der Bauer allein die Seite oder nimmt genau ein Objekt mit sich.

### Entstehender Graph
Der Graph wird ungerichtet aufgebaut, da jede Fahrt durch eine Fahrt mit gleicher Ladung in die entgegengesetzte Richtung rückgängig gemacht werden kann, sodass von symmetrischen Kanten ausgegangen werden kann. Durch die Einschränkung auf ausschließlich gültige Zustände beginnt der Graph zwangsweise wie folgt:

```
BKSW|  --  KW|BS  --  BKW|S  --  ...
```

Sobald der Graph vollständig ist, kann systematisch und vor allem effizient nach einem Pfad vom Start- zum Zielknoten gesucht werden. Falls es mehrere Wege zum Ziel gibt, ist dem Bauern mit Sicherheit der kürzeste Pfad am liebsten.
