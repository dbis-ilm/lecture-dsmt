## Distanzfunktionen
Um Objekte nach Ähnlichkeit zu gruppieren muss diese Ähnlichkeit aber natürlich für einen Algorithmus erfassbar darstellbar sein. Dafür ist lediglich eine Funktion von Nöten, die für zwei gegebene Daten einen Abstand berechnet. Relevant ist ausschließlich, dass Paaren ähnlicher Objekte geringe und Paaren unähnlicher Objekte große Distanzen zugeordnet werden.

An ein solches Distanzmaß $dist$ mit zwei Objekten $o_1, o_2$ werden daher die folgenden Anforderungen gestellt:
1. Das Ergebnis muss eine reelle Zahl sein: $$dist(o_1, o_2) = d \in \mathbb{R}$$
2. Der Abstand ist $0$, gdw. beide Objekte gleich sind. $$dist(o_1, o_2) = 0 \longleftrightarrow o_1 = o_2$$
3. Der Wert ist unabhängig der Reihenfolge der Objekte. (Symmetrie) $$dist(o_1, o_2) = dist(o_2, o_1)$$

Es muss sich dabei nicht zwangsweise um die euklidische Distanz oder ein verwandtes Maß handeln, auch wenn diese natürlich leicht verständlich und insbesondere in kleiner Dimensionalität visuell nachvollziehbar sind. Nachfolgend sollen einige Distanzfunktionen (inklusive der euklidischen Distanz) kurz gezeigt werden.

Seien dazu zwei Objekte $x = (x_1, \dots, x_n)$ und $y = (y_1, \dots, y_n)$ anhand ihrer Koordinaten gegeben. Als Basis dient die sogenannte **Minkowski-Distanz**:

$$
dist(x, y) = \sqrt[p]{\sum_{i=1}^{d} (x_i - y_i)^p}
$$

- **Euklidische Distanz:** Lassen sich die relevanten Merkmale in einem kartesischen Koordinatensystem darstellen, genügt häufig die euklidische Distanz. Für $p=2$ entsteht die bekannte Funktion: $$
dist_E(x, y) = \sqrt{(x_1 - y_1)^2 + \dots + (x_n - y_n)^2}
$$
- **Manhattan-Distanz:** Die Manhattan-Distanz erhält ihren Namen vom Layout des New Yorker Stadtteils. Für $p=1$ ergibt sich: $$
dist_M(x, y) = \sum_{i=1}^{d} \left| x_i - y_i \right|
$$
- **Maximums-Metrik:** Die Maximums-Metrik ist der Grenzwert für $p \rightarrow \inf$: $$
dist_{max} = max(\{ |x_i - y_i| : 1 \leq i \leq d \})
$$
